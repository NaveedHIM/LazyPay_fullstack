```python
lst = [{1,2},3,9,[(5),[6,7]]]

res = []


while lst:
    c = lst.pop()
    if type(c) == list or type(c) == tuple or type(c) == set:
        for i in (c):
            lst.append(i)
    else:
        res.append(c)

print(sorted(res))
```

```python
sentence = "this a that this a what"
words = sentence.split()
freq = {}

for i in words:
    if i in freq:
        freq[i]+=1
    else:
        freq[i] = 1

print(freq)
```
